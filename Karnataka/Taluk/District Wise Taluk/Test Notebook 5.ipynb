```python
gdf2.columns=['d_id', 'd_name']
for i in range(len(gdf2)):
    gdf2.iloc[i,0] = int(gdf2.iloc[i,0])
gdf2
```

```python
gdf1 = gdf[["KGISTalukN", "KGISDistri", "geometry"]]
gdf1
```

```python
for i in range(len(gdf1)):
    gdf1.iloc[i,1] = int(gdf1.iloc[i,1])
    
gdf1
```

```python
gdf1.columns = ['sd_name', 'd_id', 'geometry']
gdf = pd.merge(gdf1, gdf2, on='d_id')
gdf
```

```python
col=gdf.pop('d_name')
gdf.insert(2, 'd_name', col)
```

```python
gdf['sd_id'] = gdf.index+1
```

```python
col = gdf.pop('sd_id')
gdf.insert(0, 'sd_id', col)
```

```python
districtNames = list(gdf['d_name'].unique())
for dist in districtNames:
    df = gdf[gdf['d_name']==dist]
    path = rf"C:\Users\skeda\OneDrive\Desktop\CEGIS\Project 6\Final Shapefiles\Karnataka\Untitled Folder\{dist}"
    if not os.path.exists(path):
        os.makedirs(path)
    df.to_file(rf"C:\Users\skeda\OneDrive\Desktop\CEGIS\Project 6\Final Shapefiles\Karnataka\Untitled Folder\{dist}\{dist}.shp")
```
